### formula

$$
\begin{split}
&\ell=\text{LogSigmoid}(x)=-\log\frac{1}{1+\exp(-x)}\\
&\ell=\text{ReLU}(1-x)
\end{split}
$$

```python
# Define the logits range
x = np.linspace(-5, 5, 400)

# Define the log-sigmoid function
def logsigmoid(x):
    return np.log(1 + np.exp(-x))

# Define the ReLU function applied to (1 - x)
def relu_one_minus(x):
    return np.maximum(0, 1 - x)

# Compute the function values
y_logsigmoid = logsigmoid(x)
y_relu = relu_one_minus(x)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot log-sigmoid
plt.plot(x, y_logsigmoid, label='Log-Sigmoid(logits)', color='blue')

# Plot ReLU(1 - logits)
plt.plot(x, y_relu, label='ReLU(1 - logits)', color='red', linestyle='--')
# Add title and labels
plt.title('Plot of -Log-Sigmoid and ReLU(1 - logits) Functions')
plt.xlabel('Logits (x)')
plt.ylabel('Function Value')
```
